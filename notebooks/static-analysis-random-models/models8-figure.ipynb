```python
# read E8
#open
with open('a8_rand_results.pkl', 'rb') as file:
      
    # Call load method to deserialze
    myvar = pickle.load(file)
  
    print(myvar)
test_A8_all_agents_terms_together_thetas, test_A8_all_agents_terms_together_bal_dens, test_A8_all_agents_terms_together_bal_dens_rands = myvar
```

```python
# read C8
#open
with open('notPCA_C8_rand_results.pkl', 'rb') as file:
      
    # Call load method to deserialze
    myvar = pickle.load(file)
  
    print(myvar)
test_C8_all_agents_terms_together_thetas, test_C8_all_agents_terms_together_bal_dens, test_C8_all_agents_terms_together_bal_dens_rands = myvar
```

```python
distances = np.array([x/16 for x in range(0,17)])
pds = np.array([test_E8_all_agents_terms_together_bal_dens[np.where(test_E8_all_agents_terms_together_thetas >= distances[i])[0][0]] for i in range(0, len(distances) - 1)])

def get_mean_vals(thetas, bal_dens_rands, distances):
    mean_vals = np.mean(np.array(bal_dens_rands).T,1)
    std_vals = np.std(np.array(bal_dens_rands).T,1)
    
    pms = np.array([mean_vals[np.where(thetas >= distances[i])[0][0]] for i in range(0, len(distances) - 1)])
    stds = np.array([std_vals[np.where(thetas >= distances[i])[0][0]] for i in range(0, len(distances) - 1)]) 
    
    return pms, stds

e8_pms, e8_stds = get_mean_vals(test_E8_all_agents_terms_together_thetas, test_E8_all_agents_terms_together_bal_dens_rands, distances)

a8_pms, a8_stds = get_mean_vals(test_A8_all_agents_terms_together_thetas, test_A8_all_agents_terms_together_bal_dens_rands, distances)

c8_pms, c8_stds = get_mean_vals(test_C8_all_agents_terms_together_thetas, test_C8_all_agents_terms_together_bal_dens_rands, distances)
```

```python
fig, ax = plt.subplots()


chosen_dists = [i/8 for i in range(0,8)]
chosen_dists = [i/16 for i in range(0,16)]

anno_dists = [0, 0.125, 3/16, 4/16, 5/16, 6/16, 0.5, 9/16, 10/16, 11/16, 12/16, 13/16, 14/16]


for i, dist in enumerate(distances):
    if dist not in chosen_dists:
        continue
    pms = e8_pms
    stds = e8_stds
    dif = pds[i] - pms[i]
    ax.errorbar(pds[i], dif, yerr=stds[i], color='blue', marker='o', linestyle='', markersize=10, label=str(dist))
#     plt.errorbar(x=pds[i], y=dif,yerr=stds[i],  color='blue', marker='o', markersize=10, label=str(dist))
    if dist == 0.75:
        xy = (pds[i] + 0.02, dif-0.01)
    elif dist in [7/8]:
        xy = (pds[i]-0.00, dif + 0.0)
    elif dist in [6/16]:
        xy = (pds[i] + 0.01, dif + 0.02)
    elif dist in [8/16]:
        xy = (pds[i] + 0.01, dif - 0.02)
    elif dist in [1/16, 3/16]:
        xy = (pds[i]-0.11, dif + 0.01)
    elif dist in [11/16]:
        xy = (pds[i]-0.06, dif + 0.01)
    elif dist in [5/16]:
        xy = (pds[i]-0.15, dif + 0.01)
    elif dist in [0, 0.25]:
        xy = (pds[i]-0.07, dif - 0.03)
    elif dist in [0.625]:
        xy = (pds[i]-0.15, dif - 0.03)
#     elif dist == 0:
#         xy = (pds[i]-0.07, dif - 0.03)
    else:
        xy = (pds[i], dif)
    if dist in anno_dists:
        ax.annotate(str(dist),xy=xy)
    #, xytext=(-40,-10), textcoords='offset points')

#plt.rcParams.update({'font.size': 40})
plt.title('(d) Model E8')
fig.set_size_inches(6, 5)
plt.rcParams.update({'font.size': 15})
plt.plot((0,1),(0,0), color='black')


# plt.savefig("modelE8v2.pdf",dpi=600,bbox_inches='tight',format='pdf')

# with open(f'E8.pkl','wb') as fid:
#     pickle.dump(fig, fid)
```
